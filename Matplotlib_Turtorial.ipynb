```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Bar Plot

```python
x = ["python", "c", "c++", "java"]
y = [85, 70, 60, 82]
plt.xlabel("language", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Popularity of Languages")
c = ["y", "b", "m","g"]
plt.bar(x,y, width=0.4, color=c, align="edge", edgecolor="r", linewidth=5, linestyle=":", alpha=0.4, label="Popularity")
plt.legend()
plt.show()
```
