```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Bar Plot

```python
x = ["python", "c", "c++", "java"]
y = [85, 70, 60, 82]
plt.xlabel("language", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Popularity of Languages")
c = ["y", "b", "m","g"]
plt.bar(x,y, width=0.4, color=c, align="edge", edgecolor="r", linewidth=5, linestyle=":", alpha=0.4, label="Popularity")
plt.legend()
plt.show()
```

```python
# Multiple Bar graphs
z = [20, 30, 40, 50]
plt.xlabel("language", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Popularity of Languages", fontsize=15)
width = 0.2
plt.bar(x,y, width, color="r", label = "popularity")
plt.bar(x,z, width, color="y", label = "popularity1")
plt.legend()
plt.show()
```

```python
# To make them side by side

p = np.arange(len(x))
width = 0.2
p1 = p + width

plt.xlabel("language", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Popularity of Languages", fontsize=15)
plt.bar(p,y, width, color="r", label = "popularity")
plt.bar(p1,z, width, color="y", label = "popularity1")

plt.xticks(p + width/2, x, rotation=20)
plt.legend()
plt.show()
```

```python
# Horizontal bar graph

plt.ylabel("No.", fontsize=15)
plt.title("Popularity of Languages", fontsize=15)
plt.barh(x,y, width, color="r", label = "popularity")
plt.barh(x,z, width, color="y", label = "popularity1")

plt.legend()
plt.show()
```

# Scatter Plot

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
colors = ["r", "y", "g", "b", "r", "g", "r"]
sizes = [220, 102, 204, 115, 212, 110, 117]

plt.scatter(day,no, color=colors, s=sizes, alpha=0.4, marker="*", edgecolor="g", linewidth=2)

plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```

```python
# Color Ranges
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
no2 = [3, 2, 4, 5, 1, 6, 2]
colors = [20, 49, 30, 29, 56, 20, 30]
sizes = [220, 102, 204, 115, 212, 110, 117]

plt.scatter(day,no, c=colors, s=sizes, cmap="BrBG", alpha=0.4, label="No1")
plt.scatter(day, no2, color="b", s=sizes, alpha=0.5, label="No.2")

plt.legend()
t = plt.colorbar()
t.set_label("Colorbar", fontsize=15)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```
